## 📌 목표

추천 알고리즘에서 사용할 데이터셋을 생성한다.

### 진행 내용

- TeamMatchTbl 불러오기
- Champion ID → Champion Name 변환
- Blue Team 데이터 생성
- Red Team 데이터 생성
- 하나의 Team Dataset 생성

```python
import pandas as pd

# 데이터 불러오기
team = pd.read_csv("../data/raw/TeamMatchTbl.csv")
champion = pd.read_csv("../data/raw/ChampionTbl.csv")
```

```python
# Champion ID → Champion Name 변환
champion_dict = dict(zip(champion["ChampionId"], champion["ChampionName"]))

for col in [
    "B1Champ","B2Champ","B3Champ","B4Champ","B5Champ",
    "R1Champ","R2Champ","R3Champ","R4Champ","R5Champ"
]:
    team[col] = team[col].map(champion_dict)
```

```python
team.head()
```

### 결과

모든 챔피언 ID를 실제 챔피언 이름으로 변환하였다.

```python
blue_team = team[
    ["B1Champ","B2Champ","B3Champ","B4Champ","B5Champ","BlueWin"]
].copy()
```

```python
blue_team.columns = [
    "Champ1",
    "Champ2",
    "Champ3",
    "Champ4",
    "Champ5",
    "Win"
]
```

```python
red_team = team[
    ["R1Champ","R2Champ","R3Champ","R4Champ","R5Champ","RedWin"]
].copy()
```

```python
red_team.columns = [
    "Champ1",
    "Champ2",
    "Champ3",
    "Champ4",
    "Champ5",
    "Win"
]
```

두 데이터 합치기 -> concat

```python
team_dataset = pd.concat(
    [blue_team, red_team],
    ignore_index=True
)
```

```python
team_dataset.head()
```

팀 조합을 하나의 문자열로 만들기 

```python
team_dataset["TeamComp"] = team_dataset[
    ["Champ1", "Champ2", "Champ3", "Champ4", "Champ5"]
].apply(lambda x: ",".join(sorted(x)), axis=1)
```
